```python
#import pandas and numpy
import pandas as pd
import numpy as np
```

```python
#load data
url = (     "https://raw.github.com/pandas-dev"
            "/pandas/master/pandas/tests/io/data/csv/tips.csv"
    )
tips = pd.read_csv(url)
tips
```

# Select

```python
# SELECT total_bill, tip, smoker, time
# FROM tips;
tips[["total_bill", "tip", "smoker", "time"]]
```

# Append new coulmn

```python
# SELECT *, tip/total_bill as tip_rate
# FROM tips;
# DataFrame.assign() to append a new column
tips.assign(tip_rate=tips["tip"] / tips["total_bill"])
```

# Where

```python
# SELECT *
# FROM tips
# WHERE time = 'Dinner';
tips[tips["total_bill"] > 10]
```

# Value counts

```python
# count the values in the table
is_dinner = tips["time"] == "Dinner"
is_dinner.value_counts()
```

```python
#change the column name
tips[is_dinner]
```

# And

```python
# SELECT *
# FROM tips
# WHERE time = 'Dinner' AND tip > 5.00;
tips[(tips["time"] == "Dinner") & (tips["tip"] > 5.00)]
```

# OR

```python
# SELECT *
# FROM tips
# WHERE size >= 5 OR total_bill > 45;
tips[(tips["size"] >= 5) | (tips["total_bill"] > 45)]
```

# Groupby

```python
# SELECT sex, count(*)
# FROM tips
# GROUP BY sex;
tips.groupby("sex").size()
```

```python
# counts all the column result
tips.groupby("sex").count()
```

```python
# count the specific column
tips.groupby("sex")["total_bill"].count()
```

# Average

```python
# SELECT day, AVG(tip), COUNT(*)
# FROM tips
# GROUP BY day;

tips.groupby("day").agg({"total_bill":np.mean,"day":np.size})
```

```python
# SELECT smoker, day, COUNT(*), AVG(tip)
# FROM tips
# GROUP BY smoker, day;

tips.groupby(["smoker","day"]).agg({"total_bill":np.mean,"day":np.size})
```

# Join

```python
df1 = pd.DataFrame({"key": ["A", "B", "C", "D"], "value": np.random.randn(4)})
df2 = pd.DataFrame({"key": ["B", "D", "D", "E"], "value": np.random.randn(4)})
```

# Inner Join

```python
# SELECT *
# FROM df1
# INNER JOIN df2
#   ON df1.key = df2.key;
pd.merge(df1, df2, on="key")
```

```python
# new table with specific column
indexed_df2 = df2.set_index("key")
pd.merge(df1, indexed_df2, left_on="key", right_index=True)
```

# LEFT OUTER JOIN

```python
# SELECT *
# FROM df1
# LEFT OUTER JOIN df2
#   ON df1.key = df2.key;
pd.merge(df1, df2, on="key", how="left")
```

# RIGHT JOIN

```python
# SELECT *
# FROM df1
# RIGHT OUTER JOIN df2
#   ON df1.key = df2.key;
pd.merge(df1, df2, on="key", how="right")
```

# FULL JOIN

```python
# SELECT *
# FROM df1
# FULL OUTER JOIN df2
#   ON df1.key = df2.key;
pd.merge(df1, df2, on="key", how="outer")
```

# UNION

```python
#concat(merge two table)
df1 = pd.DataFrame( {"city": ["Chicago", "San Francisco", "New York City"], "rank": range(1, 4)})
df2 = pd.DataFrame({"city": ["Chicago", "Boston", "Los Angeles"], "rank": [1, 4, 5]})
```

```python
# SELECT city, rank
# FROM df1
# UNION ALL
# SELECT city, rank
# FROM df2;
pd.concat([df1, df2])
```

```python
# SELECT city, rank
# FROM df1
# UNION
# SELECT city, rank
# FROM df2;
pd.concat([df1, df2]).drop_duplicates()
```

# Offset

```python
# SELECT * FROM tips
# ORDER BY tip DESC
# LIMIT 10 OFFSET 5;
tips.nlargest(10 + 5, columns="tip").tail(10)
```

```python
# -- Oracle's ROW_NUMBER() analytic function
# SELECT * FROM (
#   SELECT
#     t.*,
#     ROW_NUMBER() OVER(PARTITION BY day ORDER BY total_bill DESC) AS rn
#   FROM tips t
# )
# WHERE rn < 3
# ORDER BY day, rn;

tips.assign(rn=tips.sort_values(["total_bill"], ascending=False).groupby(["day"]).cumcount()+1).query("rn < 3").sort_values(["day", "rn"])
```

# Update

```python
# UPDATE tips
# SET tip = tip*2
# WHERE tip < 2;
tips.loc[tips["tip"] < 2, "tip"] *= 10
tips.nlargest(10 + 5, columns="tip")
```
