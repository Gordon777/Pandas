```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'login':[11, 12, 13, 14, 15, 16, 11, 12, 12, 13],
    'deposit':[100, 0, 0, 400, 0, 0, 0, 200, 300, 0],
    'comment':['$\str_11', '42/str_12', '89str_13', '8str_14', 'str_15', 'str_16', 'str_11ui', 'str_12.', 'str_12$%', 'str_13']
})
df
```

```python
df_name = pd.DataFrame({
    'login':[12,13,14,15,16,17,18],
    'name':['name_12','name_13','name_14','name_15','name_16','name_17','name_18']
})
df_name
```

```python
#清理comment，改成str_xx
df.drop_duplicates(subset=['login'])
sr = df['comment']
idx = df['login']
sr.index = idx
df_comment = sr.str.extract(pat = '(str_..)')
df_comment
```

```python
#根據login，加總deposit、計算login的個數
df['deposit_sum'] = df.groupby('login')['deposit'].transform('sum')
df['login_count'] = df.groupby('login')['login'].transform('count')
df2 = df.drop_duplicates(subset=['login'])
df2
```

```python
#合併df_name
df_sum = pd.merge(df2,df_comment,how = 'left',on='login')
df_all = pd.merge(df_sum,df_name,how = 'left',on='login')
df5 = df_all.drop_duplicates(subset=['login'])
df5.drop(columns=['comment'])
```
